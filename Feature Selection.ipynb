# How to Use Feature Importance
Feature importance refers to technique that assign a score to input features based on how useful they are at predecting a target variable. There are many types and sources of feature importance scores, alsthough popular examples include statistical correlation scores, coefficients claculated as part of linear models, decision trees, and permutation importnace scores.

### Feautre Importance
It is usefult to :
1. Better Understand The data.
2. Better Understand the model.
3. Reducing the number of input features.

## Test Datasets
We will use both classification and regression datasets

## Coeffficients as Feature Importance
Linear machine learning algorithms fit a model where the prediction is the wrighted sum of the input values. Examples include linear regression, logistic regression, and extension that add regularization, such as redge regression, LASSO, and the Elastic net. all of these algorithm find find a set of coefficients to use in the wrighted sum in order to make a predction. THese coefficient can be used directly as a crude type of feature importance socre.

#### Linear Regression Feature Importance

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# define the dataset 
X, y = make_regression(n_samples = 1000, n_features=10, n_informative=5, random_state=1)

#Define the model
model = LinearRegression()
model.fit(X,y)
# get importance
importance= model.coef_
for i, v in enumerate(importance):
    print(f"Feature: {i:.3f}, Score: {v:.3f}")

plt.bar([x for x in range(len(importance))], importance)
plt.grid(True)
plt.show()
```

#### Logistic Regression Feature Importance

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
X,y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, random_state=1)
model_lor = LogisticRegression()
model_lor.fit(X, y)
importance_lor = model_lor.coef_[0]
for i, v in enumerate(importance_lor):
    print(f"Feature: {i:.3f}, Score:{v:.3f}")

plt.bar([x for x in range(len(importance_lor))], importance_lor)
plt.grid(True)
plt.show()
```

### Decision Tree Feautre Importance

#### CART Feature Importance

```python
from sklearn.datasets import make_regression 
from sklearn.tree import DecisionTreeRegressor

X,y = make_regression(n_samples=1000, n_features=10, n_informative=5, random_state=1)

model_dtr= DecisionTreeRegressor()
model_dtr.fit(X,y)
importance_dtr = model_dtr.feature_importances_
for i, v in enumerate(importance_dtr):
    print(f"Feature: {i}, Score: {v}")
plt.bar([x for x in range(len(importance_dtr))], importance_dtr)
plt.grid(True)
plt.show()
```

#### CART Classification Feature Importance

### Permutation Feature Importance
Permutation feature importnace is a texhnique for claculating relative importnce socres that is indenpendent of the model used. First, a model is fit on the dataset, such as a modle that does not support native feature importance socres. Then the model is used to make predictions on a dataset. although the vlaues of a feature(column) in the datset are scrambled.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.inspection import permutation_importance

model_pfir = KNeighborsRegressor()
model_pfir.fit(x,y)
results= permutation_importance(model_pfir, x, y, scoring='neg_mean_squared_error')
imp_pfir = results.importances_mean

for i, v in enumerate(imp_pfir):
    print(f"Feature:{i:.3f}, Scoring: {v:.3f}")

plt.bar([x for x in range(len(imp_pfir))], imp_pfir)
plt.show()
```

```python
### Permutation feature Importance Classification 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import permutation_importance

model_pfic = KNeighborsClassifier()
model_pfic.fit(a,b)
results = permutation_importance(model_pfic, a, b, scoring='accuracy')
# get importance
imp_pfic= results.importances_mean
for i,v in enumerate(imp_pfic):
    print(f"Feature:{i:.3f}, scores: {v:.3f}")

plt.bar([x for x in range(len(imp_pfic))], imp_pfic)
plt.show()
```

### Feature Selection With Importance

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
#Define dataset
def load_dataset():
    x, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, random_state=1)
    return x, y

def select_features(x_train, y_train, x_test):
    fs = SelectFromModel(RandomForestClassifier(n_estimators=1000), max_features=5)
    fs.fit(x_train, y_train)
    x_train_fs = fs.transform(x_train)
    x_test_fs = fs.transform(x_test)
    return x_train_fs, x_test_fs, fs

x, y = load_dataset()
# split them into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.33, random_state=1)
# select features
x_train_fs , x_test_fs, fs = select_features(x_train, y_train, x_test)
# train the model 
model = LogisticRegression(solver='liblinear')
model.fit(x_train_fs, y_train)

pred = model.predict(x_test_fs)

# Now we will check the score
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy*100:.3f}")
```
